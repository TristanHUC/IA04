```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# read data from beer_sales_log.txt
df = pd.read_csv('beer_sales_log_10barmen.txt', sep='\t', header=None)
df.head()
```

```python
# each line is a transaction, the only column is the time of the transaction
# create a new column to indicate the number of transactions
df['count'] = 1
df.head()
```

```python
# create a new column to indicate cumulated number of transactions
df['cum_count'] = df['count'].cumsum()
df.head()
```

```python
# rename the first column to "tick"
df.rename(columns={0: 'tick'}, inplace=True)
df.head()
```

```python
 # plot the cumulated number of transactions
sns.lineplot(x='tick', y='cum_count', data=df)
```

```python
# same for 3 barmen
df2 = pd.read_csv('beer_sales_log_3barmen.txt', sep='\t', header=None)
df2['count'] = 1
df2['cum_count'] = df2['count'].cumsum()
df2.rename(columns={0: 'tick'}, inplace=True)
df2.head()
```

```python
# plot the two on the same graph
sns.lineplot(x='tick', y='cum_count', data=df, label='10 barmen')
sns.lineplot(x='tick', y='cum_count', data=df2, label='3 barmen')
# set y and x axis labels
plt.ylabel('cumulated number of beers sold')
```
